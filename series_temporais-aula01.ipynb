# List comprehension

```python
lista = []
for valor in range(5):
    lista.append(valor+10)
print(lista)
```

```python
lista = [valor+10 for valor in range(5)]
print(lista)
```

```python
lista = []
for numero in range(1,30):
    if numero % 4 == 0:
        lista.append(numero)
        
print(lista)
```

```python
lista = [numero for numero in range(1,30) if numero % 4 == 0]
print(lista)
```

```python
conceito = ['azul' if  nota>= 6 else 'vermelha' for nota in range(1,11)]
print(conceito)
```

# Teste de estacionariedade

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams

rcParams['figure.figsize'] = 15, 6

```

```python
np.random.seed(10)
dados1 = np.random.normal(0,1,41)



dados1 = pd.DataFrame(dados1)
dados1.columns = ['valores']
# dados1


indice = pd.date_range('1980', periods = len(dados1), freq = 'Y')
# indice


serie1 = pd.Series(dados1['valores'].values, index = indice)
serie1
```

```python
serie1.plot()
plt.show()
```

```python
import statsmodels.tsa.stattools
```

# Teste KPSS

Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

Ha = não é estacionária: estatística do teste > valor crítico.

Ho = é estacionária:  estatística do teste < valor crítico.

```python
import statsmodels.tsa.stattools

kpss = statsmodels.tsa.stattools.kpss(serie1)

print("Estatistica do teste: {:.4f} ".format(kpss[0]))
print("p-valor: {} ".format(kpss[1]))
print("# Lags {}".format(kpss[2]))

print("Valores Críticos: ")
for chave, valor in kpss[3].items():
    print("\t\tchave {} \t valor {} ".format(chave, valor))
```

# Teste Dickey Fuller

Teste df (Dickey Fuller)

Ho = não é estacionária: estatística do teste > valor crítico.

Ha = é estacionária:  estatística do teste < valor crítico.

```python
from statsmodels.tsa.stattools import adfuller

dickeyFuller = adfuller(serie1)

print("Estatistica do teste: {:.4f} ".format(dickeyFuller[0]))
print("p-valor: {} ".format(dickeyFuller[1]))
# print("# Lags {}".format(dickeyFuller[2]))

print("Valores Críticos: ")
for chave, valor in dickeyFuller[4].items():
    print("\t\tchave {} \t valor {} ".format(chave, valor))
```

```python
## Série Passeio aleatório
serie2 = list()
serie2.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movimento = -1 if random() < 0.5 else 1
    valor = serie2[i-1] + movimento
    serie2.append(valor)
```

```python
random()
```

```python
serie2 = pd.Series(serie2)
serie2.plot()
plt.show()
```
